```python
import matplotlib.pyplot as plt

from helpers.plots import plot_bifurcation
from helpers.math import XDotPoly
```

## 1. Plot of saddle-node bifurcation

```python
ALPHA_MIN = -2
ALPHA_MAX = 2
BIFURCATION_POINT = 0
P_FUNC = lambda alpha: [-1, 0, alpha]  # lambda alpha: [-2, 0, alpha - 3]
TITLE = r"Bifurcation diagram of saddle-node bifurcation $\dot{x} = \alpha - x^2$"

alphas, x, stability = XDotPoly.get_stability_for_parametric(P_FUNC, ALPHA_MIN, ALPHA_MAX)

fig, ax = plt.subplots(1, 1, figsize=(7, 7))
plot_bifurcation(alphas, x, stability, ax=ax)
ax.axvline(x=BIFURCATION_POINT, color="k", label="bifurcation point")
ax.legend()
ax.set_xlim([ALPHA_MIN, ALPHA_MAX])
ax.set_xlabel(r"$\alpha$")
ax.set_ylabel(r"$x$")
ax.set_title(TITLE)
ax.grid()
fig.tight_layout()
fig.savefig("plots/task_2_bifurcation_diagram_1.pdf")
plt.show()
```
